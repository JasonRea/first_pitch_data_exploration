```python
import pandas as pd
import numpy as np
```

```python
# combine data frames of first pitch 9th inning data with first pitch data for balls out of the strike zone

# Clean and join data
first_pitch_data = pd.read_csv("../data/ninth_inning_first_pitch_data.csv")
first_pitch_data = first_pitch_data[first_pitch_data['pitches'] >= 100] # Include only pitchers who faced 100 batters in the 9th
first_pitch_data = first_pitch_data[['pitches', 'player_id', 'player_name', 'swings', 'takes']]

out_of_zone_data = pd.read_csv("../data/out_of_zone_data.csv")
out_of_zone_data = out_of_zone_data[['pitches_out_of_zone', 'player_id', 'player_name']]

data = pd.merge(first_pitch_data, out_of_zone_data, on=['player_id', 'player_name'], how='outer')
data = data[pd.notna(data['pitches'])]

# Calculate first out of zone percentage
data['first_pitch_ooz_pct'] = data['pitches_out_of_zone'] / data['pitches']

# Sort by out of zone percent
data = data.sort_values('first_pitch_ooz_pct', ascending=False)

data.head()
```
